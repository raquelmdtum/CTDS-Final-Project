```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
import plotly.graph_objects as go
import plotly.offline as pyo
import plotly.io as pio
import seaborn as sns
```

```python
products = pd.read_csv('data/skincare.csv')
```

```python
products.columns
```

```python
products.head(2)
```

```python
def counts(column, column_name, top_10_df = True, top_10_plots = True):

    count = products[column].value_counts()
    percentage = products[column].value_counts(normalize=True)*100

    df = pd.DataFrame({column_name:count.index,'Counts':count.values,'Percent':np.round(percentage.values,2)})

    if top_10_df == True:
        print(df[:10])
    
    if top_10_plots == True:
        plt.figure(figsize=(8,4))
        sns.barplot(df[:10], x=column_name, y="Counts")
        plt.xticks(rotation=90)
        plt.ylabel('Count')
        plt.title(f'Top 10 Occuring {column_name}s')
        plt.show()

    return
```

```python
counts('brand_name', 'Brand')
```

```python
counts('tertiary_category', 'Product Type')
```
